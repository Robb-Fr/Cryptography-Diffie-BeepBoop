### j-invariant

 j-invariant: \\(j = 1728\frac{4a^3}{4a^3 + 27b^2} \\)

```python
def j_invariant(a,b):
    part_a = 4*(a^3)
    part_b = (4*(a^3))+(27*(b^2))
    j_inv = 1278*(part_a/part_b)
    return j_inv, part_a, part_b
```

```python
j, pa, pb = j_invariant(a,b)
```

```python
gcd_ab = gcd(pa,pb)
```
