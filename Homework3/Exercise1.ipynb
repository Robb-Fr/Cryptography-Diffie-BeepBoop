```python
def j_invariant(a,b):
    part_a = 1728 * 4 * a**3
    part_b = (4*(a**3))+(27*(b**2))
    #part_b = 4 * a**3 + 27 * b**2
    j_inv = (part_a/part_b)
    return j_inv, part_a, part_b
```
