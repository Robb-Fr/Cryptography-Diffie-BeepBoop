### j-invariant

 j-invariant: \\(j = 1728\frac{4a^3}{4a^3 + 27b^2} \\)

```python
def j_invariant(a,b):
    part_a = 1278*4*(a^3)
    part_b = (4*(a^3))+(27*(b^2))
    j_inv = (part_a/part_b)
    return j_inv, part_a, part_b
```

```python
j, pa, pb = j_invariant(sola,modsolb)
j, pa, pb
```

```python
gcd_ab = gcd(pa,pb)
```

```python
final_a = pa/gcd_ab
final_b = pb/gcd_ab
```

```python
final_a, final_b
```

```python
assert(gcd(final_a,final_b)==1)
```

```python
assert(j == pa/pb)
```
